```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Quest√£o 3 

def f(x):
  y = np.log(x) - (2 * x ) + 1 
  return y

def zero(x):
  y = x -x
  return y

def ploting():
  a = float(input ("Informe o valor de x: "))
  b = float(input ("Informe o valor de y: "))
  
  valores = np.linspace(a,b)

  plt.plot(valores, f(valores))
  plt.plot(valores, zero(valores))
  plt.xlabel("eixo X")
  plt.ylabel("eixo Y")
  plt.grid(True)

ploting()
```
