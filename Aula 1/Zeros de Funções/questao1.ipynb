```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Atividade 1 - Questão 1 
def f(x):
  y = 2*(x**3) - (5 * (x ** 2)) - x + 3 
  return y

def zero(x):
  y = x -x
  return y

def menu():
  a = float(input ("Informe o valor de x: "))
  b = float(input ("Informe o valor de y: "))
  
  valores = np.linspace(a,b)

  plt.plot(valores, f(valores))
  plt.plot(valores, zero(valores))
  plt.xlabel("eixo X")
  plt.ylabel("eixo Y")
  plt.grid(True)

menu()
```
