```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exemplo 1
def f(x):
  y = x**3 - np.cos(x)
  return y

#funcao y=0
def zero (x):
  y = x -x
  return y

# gerando o grafico da funcao f

valores = np.linspace(-2,2)

plt.plot(valores, f(valores))
plt.plot(valores, zero(valores))
plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.grid(True)
```
